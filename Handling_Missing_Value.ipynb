# Method of Handling Missing Value
1. Drop Missing  Value
2. Filling
*   Mode
*   Mean
*   Median







```python
# 1. Drop Missing Value based on column
df = df.drop(columns='Series')
```

```python
df.head()
```

```python
# check rows and columns after drop column series

df.shape
```

```python
# Filling column Category with Mode

df['Model'] = df['Model'].fillna(df['Model'].mode()[0])
df['Processor'] = df['Processor'].fillna(df['Processor'].mode()[0])
df['Processor_Gen'] = df['Processor_Gen'].fillna(df['Processor_Gen'].mode()[0])
df['Hard_Disk_Capacity'] = df['Hard_Disk_Capacity'].fillna(df['Hard_Disk_Capacity'].mode()[0])
df['OS'] = df['OS'].fillna(df['OS'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Filling RAM with Mean

df['RAM'] = df['RAM'].fillna(df['RAM'].mean())
```
